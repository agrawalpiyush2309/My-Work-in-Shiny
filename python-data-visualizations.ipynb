```python
# varieties using boxplot
fig,ax = plt.subplots(nrows = 1, ncols = 1)

varieties = list(a["variety"].unique())
colors = ['red', 'blue', 'green']
i = 1
for variety,color in zip(varieties,colors):
    b = a[a["variety"] == variety].reset_index(drop = True)
    bp = ax.boxplot(b["petal.length"].values,positions = [i],patch_artist=True)
    bp['boxes'][0].set_facecolor(color)
    i+=1

ax.set_xlim(0.5, 3.5) 
ax.set_xticks(range(1,len(varieties)+1))
ax.set_xticklabels(varieties)

plt.show()
```

```python
#

values_dict = {'Setosa': 'r', 'Versicolor': 'b', 'Virginica': 'g'}

fig,ax = plt.subplots(nrows = 4, ncols = 4, figsize=(8, 8))

fig.subplots_adjust(wspace=0, hspace=0)

variables = 4
columns = list(a)[0:-2]

grouped = a.groupby('variety')

for columnrow,i in  zip(columns,range(0,variables)):
    for columncolumn,j in zip(columns,range(0,variables)):
        ax[i,j].spines['top'].set_visible(False)
        ax[i,j].spines['right'].set_visible(False)
        if (i==j):
            if(i!=0):
                ax[i,j].set_yticks([])
            for key,group in grouped:
                ax[i,j].hist(group[columnrow], 10, normed=1, histtype='bar', stacked=True,facecolor=values_dict[key], alpha=0.75,edgecolor='black')
        else:
            if(j>0):
                ax[i,j].set_yticks([])
                if(i!=variables-1): 
                    ax[i,j].set_yticks([])
            for key, group in grouped:
                points =  ax[i,j].plot(group[columnrow],group[columncolumn],'o',markersize=2,color=values_dict[key])
        if(j==0):
            ax[i,j].set_ylabel(columnrow)
        if(i==variables-1):
            ax[i,j].set_xlabel(columncolumn)
        
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

N = 5
men_means = (20, 35, 30, 35, 27)
men_std = (2, 3, 4, 1, 2)

ind = np.arange(N)  # the x locations for the groups
width = 0.35       # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(ind, men_means, width, color='r', yerr=men_std)

women_means = (25, 32, 34, 20, 25)
women_std = (3, 5, 2, 3, 3)
rects2 = ax.bar(ind + width, women_means, width, color='y', yerr=women_std)

# add some text for labels, title and axes ticks
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(ind + width/ 2)
ax.set_xticklabels(('G1', 'G2', 'G3', 'G4', 'G5'))

ax.legend((rects1[0], rects2[0]), ('Men', 'Women'))


def autolabel(rects):
    """
    Attach a text label above each bar displaying its height
    """
    for rect in rects:
        height = rect.get_height()
        print("X is {}. Width is{}".format(rect.get_x(),rect.get_width()))
        ax.text(rect.get_x() + rect.get_width()/2., 1.05*height,
                '%d' % int(height),
                ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

plt.show()
```
